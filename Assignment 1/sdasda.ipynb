```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
import matplotlib.pyplot as plt

DataSet = '/home/shafeen/Documents/My-all-programs--/SEMESTER-3/Programming_for_AI/Assignment 1/datasets'
Data_of_category = []
Data_to_extract_category = []

# Load all the xml files and extract data from them
for filename in os.listdir(DataSet):
    if filename.endswith('.xml'):
        file_path = os.path.join(DataSet, filename)
        tree = ET.parse(file_path)
        root = tree.getroot()

        for element in root.iter('category'):
            Data_to_extract_category.append(element.text)

# Create a DataFrame with the extracted category data
FIGUREE = pd.DataFrame({'category': Data_to_extract_category})

# Count the occurrences of each category
occurrences = FIGUREE['category'].value_counts().tolist()
categories = FIGUREE['category'].unique().tolist()

# Define colors for the bar chart
colors = [
    'LightSeaGreen',
    'LightCoral',
    'Khaki',
    'Thistle',
    'LightSlateGray',
    'Aquamarine'
]

# Create the bar chart
plt.bar(categories, occurrences, color=colors)
plt.xlabel('Category')
plt.ylabel('No of Occurrences')
plt.xticks(rotation='vertical')
plt.show()
```
