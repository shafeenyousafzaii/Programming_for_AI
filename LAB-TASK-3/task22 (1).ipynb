```python
print("Welcome to the adventure of the forest!!")
print("This game is designed to send you to paths where you shall discover different kinds of")
print('Enter "Hello" to proceed to the game')
powers = ["sword", "magic", "help"]

while True:
    choice = input("Enter: ").lower()

    if choice != "hello":
        print("Invalid input. Please enter 'Hello' to start the game or 'quit' to exit.")
        continue

    print("Welcome, you are in the game!")
    print('Press "continue" to proceed!')
    user_input = input().lower()

    if user_input == "continue":
        print("Here we go!")

        while True:
            print("You are walking through the jungle and you come across a monster. He's looking at you angrily.")
            print("What are you going to do?")
            print("Sword, Magic, Help, Quit!")

            user_input = input().lower()

            if user_input in powers:
                if user_input == "sword":
                    print("You chose the sword. Here comes the mighty sword, slicing away the monster. YOU WIN!!!")
                elif user_input == "magic":
                    print("You chose magic. You cast a powerful spell and defeat the monster. YOU WIN!!!")
                elif user_input == "help":
                    print("You chose help. A friendly creature appears and helps you defeat the monster. YOU WIN!!!")
                elif user_input == "quit":
                    print("Goodbye! Thanks for playing.")
                    exit()
                    
                choice = input("Enter 'Hello' to start again or 'quit' to exit: ").lower()

                if choice == "hello":
                    break
                elif choice == "quit":
                    print("Goodbye! Thanks for playing.")
                    exit(1)
                else:
                    print("Invalid choice. Exiting the game.")
                    exit(1)
            elif user_input == "quit":
                print("Goodbye! Thanks for playing.")
                exit(1)
            else:
                print("Invalid power choice. Choose from: Sword, Magic, Help, Quit!")
    elif user_input == "quit":
        print("Goodbye! Thanks for playing.")
        exit()
    else:
        print("Invalid input. Please enter 'continue' to proceed or 'quit' to exit.")
print("Quit")
```
